```python
lista_country = [country[1] for country in country_list.countries_for_language('en')]
lista_country
```

```python
lista_location = covid_data['location'].unique().tolist()
lista_location
```

```python
# verifico se i paesi del mondo coincidano con i paesi presenti nel dataset covid_data
if lista_country == lista_location:
    print("Le liste sono uguali.")
else:
    print("Le liste sono diverse.")
```

```python
new_list = []
for x in lista_location:
    if x not in lista_country:
        new_list.append(x)
```

```python
covid_data = covid_data[~covid_data['location'].isin(new_list)]
```

```python
# cambio il formato in datetime
covid_data['date'] = pd.to_datetime(covid_data['date'])

# trovo l'ultima misurazione dei casi per continente
last_day_per_continente = covid_data.groupby('continent')['date'].max()

covid_data_last_day = covid_data[covid_data['date'].isin(last_day_per_continente)]

# raggruppo per continente e sommo i casi totali
casi_per_continente_last_day = covid_data_last_day.groupby('continent')['total_cases'].sum()

# stampo il numero totale di casi per ogni continente
print(casi_per_continente_last_day)
```

```python
continente_1 = 'Africa'
continente_2 = 'Oceania'

# totale casi per continente
total_cases_per_continente = covid_data.groupby('continent')['total_cases'].sum()

# totale casi nel mondo, incluso i casi senza indicazione di continente
world_total_cases = covid_data['total_cases'].sum()

# totali dei casi per Africa e Oceania
total_cases_continente_1 = total_cases_per_continente[continente_1]
total_cases_continente_2 = total_cases_per_continente[continente_2]

# massimo valore dei casi per l'Africa
max_value_continente_1 = round(total_cases_continente_1.max(), 2)

# massimo valore dei casi per l'Oceania
max_value_continente_2 = round(total_cases_continente_2.max(), 2)

# media dei casi per i due continenti
mean_cases_continente_1 = round(total_cases_continente_1 / len(covid_data[covid_data['continent'] == continente_1]), 2)
mean_cases_continente_2 = round(total_cases_continente_2 / len(covid_data[covid_data['continent'] == continente_2]), 2)

# trovo le percentuali dei casi rispetto al totale dei casi nel mondo
percentage_cases_continente_1 = round((total_cases_continente_1 / world_total_cases) * 100, 2)
percentage_cases_continente_2 = round((total_cases_continente_2 / world_total_cases) * 100, 2)

print('Descrittori statistici per i continenti Africa e Oceania')
print('Massimo valore di casi per Africa', ':', max_value_continente_1)
print('Massimo valore di casi per Oceania', ':', max_value_continente_2)
print('Media dei casi per Africa', ':', mean_cases_continente_1)
print('Media dei casi per Oceania', ':', mean_cases_continente_2)
print('Percentuale dei casi rispetto al totale dei casi nel mondo per Africa', ':', percentage_cases_continente_1)
print('Percentuale dei casi rispetto al totale dei casi nel mondo per Oceania', ':', percentage_cases_continente_2)
```

```python
import matplotlib.pyplot as plt

italia_2022 = covid_data[(covid_data['location'] == 'Italy') & (covid_data['date'].dt.year == 2022)]

# grafico per l'evoluzione dei casi totali
plt.figure(figsize=(10, 6))
plt.plot(italia_2022['date'], italia_2022['total_cases'])
plt.title('Evoluzione dei casi totali in Italia nel 2022', color = "green")
plt.xlabel('Data')
plt.ylabel('Casi totali')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# grafico per il numero di nuovi casi rispetto alla data
plt.figure(figsize=(10, 6))
plt.plot(italia_2022['date'], italia_2022['new_cases'])
plt.title('Numero di nuovi casi in Italia nel 2022', color = "green")
plt.xlabel('Data')
plt.ylabel('Nuovi casi')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# somma cumulativa dei nuovi casi per l'anno 2022
italia_2022['cumulative_new_cases'] = italia_2022['new_cases'].cumsum()

# andamento della somma cumulativa dei nuovi casi
plt.figure(figsize=(10, 6))
plt.plot(italia_2022['date'], italia_2022['cumulative_new_cases'])
plt.title('Somma cumulativa dei nuovi casi in Italia nel 2022', color = "green")
plt.xlabel('Data')
plt.ylabel('Somma cumulativa dei nuovi casi')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

nazioni_selezionate = ['Italy', 'Germany', 'France']
filtro_nazioni = covid_data[covid_data['location'].isin(nazioni_selezionate)]

# periodo da maggio 2022 ad aprile 2023
filtro_periodo = filtro_nazioni[(filtro_nazioni['date'] >= '2022-05-01') & (filtro_nazioni['date'] <= '2023-04-30')]

# creo il boxplot con Plotly
fig = px.box(filtro_periodo, x='location', y='icu_patients', color='location',
             labels={'location': 'Nazione', 'icu_patients': 'Numero di pazienti in terapia intensiva (ICU)'})

fig.update_layout(title_text='Distribuzione dei pazienti in terapia intensiva (ICU) da maggio 2022 ad aprile 2023',
                  title_font_color='green')
fig.show()
```

```python
nazioni_selezionate = ['Italy', 'Germany', 'France', 'Spain']
filtro_nazioni_2023 = covid_data[(covid_data['location'].isin(nazioni_selezionate)) & (covid_data['date'].dt.year == 2023)]

# somma dei pazienti ospitalizzati per ognuna di queste nazioni
somma_pazienti_ospitalizzati = filtro_nazioni_2023.groupby('location')['hosp_patients'].sum()

# stampo la somma dei pazienti ospitalizzati per ogni nazione
print("Somma dei pazienti ospitalizzati per Italia, Germania, Francia e Spagna nel 2023:")
print(somma_pazienti_ospitalizzati)

# verifico se ci sono dati nulli
if filtro_nazioni_2023['hosp_patients'].isnull().any():
    print("Ci sono dati nulli nella colonna 'hosp_patients'.")
else:
    print("Non ci sono dati nulli nella colonna 'hosp_patients'.")
```
